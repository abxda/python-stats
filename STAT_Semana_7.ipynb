## Introducción

En esta clase, exploraremos las distribuciones de variables continuas, enfocándonos en dos de las más importantes: la distribución normal y la distribución t de Student.

## Distribución Normal

### Conceptos Básicos

La distribución normal, también conocida como curva de Gauss, es una de las más utilizadas en estadística debido a su capacidad para modelar una gran variedad de fenómenos naturales y sociales. Se caracteriza por su forma de campana simétrica y se define por dos parámetros:

- **Media $(\mu)$**: Representa el centro de la distribución.
- **Desviación estándar ($\sigma$)**: Mide la dispersión de los datos alrededor de la media.

La función de densidad de probabilidad de una variable aleatoria  $ X $ con distribución normal, denotada como $N(\mu, \sigma^2)$,  está dada por:

$ f(x|\mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} $

### Propiedades

- **Simétrica:** La curva es simétrica respecto a la media.
- **Campaniforme:** La mayor parte de los datos se concentran alrededor de la media, disminuyendo la frecuencia a medida que nos alejamos de ella.
- **Asíntotas:** Las colas de la curva se extienden indefinidamente sin tocar el eje horizontal (asíntotas).

_RECORDANDO_

Una función de densidad de probabilidad (FDP) es una función que describe la probabilidad de que una variable aleatoria continua tome ciertos valores. La probabilidad de que la variable caiga dentro de un intervalo específico se calcula como el área bajo la curva de la FDP para ese intervalo.

### Tipificación y uso de tablas

Para facilitar el cálculo de probabilidades, se utiliza la **distribución normal estándar (Z)**, que tiene una media de 0 y una desviación estándar de 1,  $N(0, 1)$.  Podemos convertir cualquier variable normal a la estándar utilizando la siguiente fórmula de **tipificación**:

$ Z = \frac{X - \mu}{\sigma} $

Las **tablas de la distribución normal estándar** proporcionan la probabilidad acumulada  $ P(Z \leq z) $ para un valor dado de $ z $.

### Ejemplos Prácticos en Python con visualización

Utilizaremos la librería `scipy.stats` para trabajar con la distribución normal en Python y `matplotlib` para visualizaciones.

**Ejemplo 1: Probabilidad en una distribución normal**

Calcular $ P(X \leq 4) $ para una distribución $ N(2, 4) $.



```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la distribución
mu = 2
sigma = 2

# Valor de x
x = 4

# Calculando la probabilidad
prob = stats.norm.cdf(x, mu, sigma)
print(f"P(X <= {x}) = {prob:.4f}")

# Visualización
x_vals = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y_vals = stats.norm.pdf(x_vals, mu, sigma)
plt.plot(x_vals, y_vals)
plt.fill_between(x_vals, y_vals, where=(x_vals <= x), color='blue', alpha=0.3)
plt.title('Distribución Normal N(2, 4)')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```

En el análisis de distribuciones de probabilidad, utilizamos diferentes funciones para calcular y visualizar las probabilidades. En particular, hay dos funciones clave que se emplean con frecuencia: la función de densidad de probabilidad (PDF) y la función de distribución acumulativa (CDF). Ambas juegan roles distintos y son esenciales en diversos contextos.

### Función de Distribución Acumulativa (CDF)

La función de distribución acumulativa, o CDF por sus siglas en inglés (Cumulative Distribution Function), se utiliza para calcular la probabilidad acumulada de que una variable aleatoria tome un valor menor o igual a un valor específico. En este contexto, la CDF nos dice la probabilidad de que $X \leq 4$ para una distribución normal con media 2 y desviación estándar 2. Esta es la razón por la que utilizamos `stats.norm.cdf(x, mu, sigma)` en el código. La CDF es particularmente útil cuando estamos interesados en conocer la probabilidad acumulada hasta un punto determinado.

### Función de Densidad de Probabilidad (PDF)

Por otro lado, la función de densidad de probabilidad, o PDF por sus siglas en inglés (Probability Density Function), se utiliza para representar la densidad de probabilidad de una variable aleatoria continua en un punto específico. La PDF nos proporciona una idea de la "densidad" o "altura" de la distribución en un punto dado, aunque no nos da una probabilidad directa sino la densidad de probabilidad. En el código, usamos `stats.norm.pdf(x_vals, mu, sigma)` para obtener los valores de la densidad de probabilidad de la distribución normal en un rango de valores. Esto es esencial para visualizar la forma de la distribución y entender cómo se distribuyen las probabilidades a lo largo de los valores posibles de la variable aleatoria.

### Explicación del Código

En el código proporcionado, se siguen estos pasos:

1. **Definición de los Parámetros de la Distribución**:
   - Se define una distribución normal con media $(\mu)$ de 2 y desviación estándar ($\sigma$) de 2.

2. **Cálculo de la Probabilidad Acumulada (CDF)**:
   - Se calcula la probabilidad acumulada de que la variable aleatoria $X$ sea menor o igual a 4 usando la CDF.
   - `prob = stats.norm.cdf(x, mu, sigma)` calcula esta probabilidad acumulada y la imprime.

3. **Visualización de la Distribución (PDF)**:
   - Se generan valores de $X$ en un rango que va desde $\mu - 3\sigma$ hasta $\mu + 3\sigma$ para cubrir la mayor parte de la distribución.
   - `y_vals = stats.norm.pdf(x_vals, mu, sigma)` calcula la densidad de probabilidad para estos valores.
   - Se grafica la PDF para mostrar la forma de la distribución normal.
   - `plt.fill_between(x_vals, y_vals, where=(x_vals <= x), color='blue', alpha=0.3)` resalta la región bajo la curva hasta el punto $x = 4$, mostrando gráficamente la probabilidad acumulada hasta ese punto.

Utilizamos la CDF (`stats.norm.cdf`) cuando necesitamos calcular la probabilidad acumulada hasta un cierto valor. Por otro lado, utilizamos la PDF (`stats.norm.pdf`) para visualizar la distribución de la densidad de probabilidad a lo largo de un rango de valores. Ambas funciones son fundamentales para el análisis y visualización de distribuciones de probabilidad, proporcionando diferentes perspectivas y tipos de información sobre la variable aleatoria en estudio.

**Ejemplo 2: Encontrar el valor de X dado una probabilidad**

En una distribución $ N(0, 1) $, ¿qué valor de  $ X $ deja una probabilidad de 0.95 a la izquierda?

Para encontrar el valor de $ X $ que deja una probabilidad de 0.95 a la izquierda en una distribución $ N(0, 1) $, utilizamos la función inversa de la distribución acumulativa (percent point function, `ppf`). También graficaremos esta distribución normal y resaltaremos la región correspondiente.

En estadística, a menudo nos encontramos con la necesidad de ir "al revés" en nuestro razonamiento sobre las probabilidades. En lugar de preguntar "¿Cuál es la probabilidad de observar un valor menor o igual a X?", podemos querer saber "¿Qué valor de X deja una probabilidad específica a su izquierda en la distribución?". Para responder a esta pregunta, recurrimos a la **función de punto porcentual** (PPF), también conocida como **función cuantil**.  

Piensa en la PPF como una "traducción" entre probabilidades y valores de la variable aleatoria. Si la función de distribución acumulada (CDF) nos dice la probabilidad acumulada hasta un punto dado, la PPF nos dice qué punto en la distribución corresponde a una probabilidad acumulada específica. Es como usar un diccionario bilingüe: en un sentido, buscamos la probabilidad asociada a un valor; en el otro, buscamos el valor asociado a una probabilidad.


```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Probabilidad
prob = 0.95

# Encontrar el valor de x
x = stats.norm.ppf(prob)
print(f"El valor de X que deja una probabilidad de {prob} a la izquierda es: {x:.4f}")

# Gráfica
x_vals = np.linspace(-3, 3, 1000)
y_vals = stats.norm.pdf(x_vals)
plt.plot(x_vals, y_vals, label='N(0,1)')
plt.fill_between(x_vals, y_vals, where=(x_vals <= x), color='blue', alpha=0.3, label=f'P(X <= {x:.4f}) = {prob}')
plt.axvline(x, color='red', linestyle='--', label=f'X = {x:.4f}')
plt.title('Distribución Normal Estándar')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```


## Distribución t de Student

### Conceptos Básicos

La distribución t de Student se utiliza principalmente en inferencia estadística cuando:

- El tamaño de la muestra es pequeño (n < 30).
- La desviación estándar de la población es desconocida y se estima a partir de la muestra.

La distribución t es similar a la distribución normal estándar en el sentido de que es simétrica y centrada en 0. Sin embargo, tiene colas más pesadas, lo que significa que hay una mayor probabilidad de observar valores extremos.

### Comparación con la Distribución Normal

A medida que el tamaño de la muestra aumenta, la distribución t se aproxima a la distribución normal estándar. Esto se debe a que a medida que n aumenta, la estimación de la desviación estándar de la población se vuelve más precisa.

# Grados de Libertad

#### Concepto de Grados de Libertad

Los grados de libertad (df) se refieren a la cantidad de valores en un cálculo que son libres de variar. En el contexto de la distribución t, los grados de libertad están estrechamente relacionados con el tamaño de la muestra y se utilizan para determinar la forma de la distribución.

#### Cálculo de Grados de Libertad

Para la mayoría de las pruebas estadísticas que utilizan la distribución t, los grados de libertad se calculan como:

$ \text{df} = n - 1 $

donde $ n $ es el tamaño de la muestra. Esto se debe a que una vez que se conoce la media de la muestra, solo $ n-1 $ valores son libres de variar; el último valor está determinado por la media de la muestra y los demás valores.

#### Importancia de los Grados de Libertad

Los grados de libertad afectan la forma de la distribución t. Con pocos grados de libertad, la distribución t tiene colas más pesadas, lo que refleja una mayor incertidumbre en la estimación de la desviación estándar de la población. A medida que los grados de libertad aumentan, la distribución t se aproxima más a la distribución normal estándar, indicando una mayor precisión en la estimación.

#### Ejemplo: El café perfecto

Imagina que quieres saber si una nueva cafetera realmente produce tazas de café con la temperatura perfecta: 80 grados Celsius.  Tomas una muestra de 5 tazas y obtienes las siguientes temperaturas:

- Taza 1: 78°C
- Taza 2: 82°C
- Taza 3: 79°C
- Taza 4: 81°C

Si sabemos que la temperatura promedio debe ser de 80°C, podemos calcular fácilmente la temperatura exacta que debe tener la quinta taza para que el promedio se mantenga. En este caso, solo las primeras cuatro tazas nos dieron información "libre" para determinar si la nueva cafetera realmente está haciendo la diferencia.

¡Tenemos 4 grados de libertad (df=4)! Con esta información, podemos usar la distribución t para evaluar si la diferencia entre la temperatura promedio observada y la temperatura "perfecta" (80°C) es estadísticamente significativa o simplemente una variación aleatoria.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Grados de libertad para diferentes distribuciones t
dfs = [1, 5, 10, 30]
x = np.linspace(-5, 5, 1000)

plt.figure(figsize=(10, 6))

# Distribución normal estándar
plt.plot(x, norm.pdf(x), label='Normal Estándar', linestyle='--')

# Distribuciones t de Student con diferentes grados de libertad
for df in dfs:
    plt.plot(x, t.pdf(x, df), label=f't con {df} grados de libertad')

plt.title('Comparación de la Distribución t de Student con Diferentes Grados de Libertad')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

## Desmitificando el Nivel de Significancia: Un Viaje desde la Intuición a la Distribución t

Imagina que eres un buscador de oro en plena fiebre del oro. Tienes una máquina que te ayuda a encontrar pepitas, pero a veces, te da una señal aunque solo haya tierra. El **nivel de significancia**, representado por la letra griega α (alfa), es como ajustar la sensibilidad de tu detector de pepitas.  

**En pocas palabras:**  es el riesgo que estás dispuesto a tomar de emocionarte por una señal que solo se debe al azar, a una falsa alarma. Es como decir: "Estoy dispuesto a aceptar que me equivoque un 5% (o un 1%, o un 10%) de las veces, pero no más".

**¿Cómo se traduce esto a la distribución t de Student?**

Imagina ahora que estás estudiando si una nueva técnica de enseñanza realmente mejora las calificaciones de tus alumnos. Tomas una muestra pequeña y ves una diferencia en las calificaciones promedio. ¿Es esta diferencia real o solo una casualidad debido a la muestra pequeña?

Aquí entra la distribución t: es como un mapa que te muestra la probabilidad de observar diferentes valores del "estadístico t" si la nueva técnica no tuviera ningún efecto real (es decir, si solo estuvieras encontrando "tierra" en lugar de "pepitas").

**El nivel de significancia te ayuda a leer el mapa:**

* **Si α = 0.05:** Trazas una línea en tu mapa que deja el 5% más extremo de la distribución t. Si tu estadístico t cae en esa zona, es como si tu detector de oro te diera una señal fuerte. Es poco probable que la diferencia en las calificaciones sea solo por casualidad. ¡Podrías tener una pepita de oro (un descubrimiento)!
* **Diferentes niveles de α:** Cambiar α es como ajustar la sensibilidad de tu detector. Un α más pequeño (por ejemplo, 0.01) significa que eres más cauteloso y solo te emocionas con señales muy fuertes, mientras que un α más grande (por ejemplo, 0.10) significa que estás dispuesto a correr más riesgos de encontrar falsas alarmas.

**Finalmente:** El nivel de significancia te ayuda a decidir si una diferencia observada es lo suficientemente grande como para ser considerada un hallazgo real en lugar de una mera casualidad. En el contexto de la distribución t,  es una herramienta esencial para interpretar tus resultados y tomar decisiones informadas.

### Uso de Python para Reemplazar las Tablas

Utilizar Python para calcular los valores críticos de la distribución t es eficiente y preciso. A continuación, se muestra cómo hacerlo utilizando la biblioteca `scipy`.

**Ejemplo 1: Probabilidad en una distribución t**

Calcular $ P(t \leq 1.3009) $ para  $ df = 5 $.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Grados de libertad
df = 5

# Valor de t
t = 1.3009

# Calculando la probabilidad
prob = stats.t.cdf(t, df)
print(f"P(t <= {t}) con df = {df} es: {prob:.4f}")

# Visualización
x_vals = np.linspace(-4, 4, 100)
y_vals = stats.t.pdf(x_vals, df)
plt.plot(x_vals, y_vals, label=f't ({df})')
plt.fill_between(x_vals, y_vals, where=(x_vals <= t), color='blue', alpha=0.3)
plt.title('Distribución t de Student')
plt.xlabel('t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```


Este código calcula el valor crítico de t para una prueba de dos colas con 10 grados de libertad y un nivel de significancia de 0.05.

### Correspondencia entre Tablas y Python

La correspondencia entre las tablas tradicionales y los cálculos en Python es directa. Las tablas proporcionan valores críticos específicos para varios grados de libertad y niveles de significancia, mientras que Python permite calcular estos valores de manera dinámica y precisa, adaptándose a cualquier combinación de grados de libertad y niveles de significancia.

- **Python**: Ejecutando el código anterior, obtienes el mismo valor crítico de 2.228.

### Ventajas del Uso de Python

- **Precisión**: Python proporciona cálculos precisos basados en fórmulas matemáticas exactas.
- **Flexibilidad**: Puedes calcular valores críticos para cualquier combinación de grados de libertad y niveles de significancia sin limitaciones de las tablas predefinidas.
- **Eficiencia**: Ahorra tiempo y reduce errores manuales en la consulta y interpretación de tablas.

En resumen, la utilización de Python no solo facilita los cálculos relacionados con la distribución t de Student, sino que también permite una mayor precisión y flexibilidad en la inferencia estadística, superando las limitaciones de las tablas tradicionales.

**Ejemplo 2: Encontrar el valor crítico de t**

En una distribución t con  $ df = 10 $, ¿cuál es el valor crítico de $ t $ que deja una probabilidad de 0.025 en la cola derecha?

```python
import scipy.stats as stats

# Grados de libertad
df = 10

# Probabilidad en la cola derecha
prob_cola_derecha = 0.025

# Encontrar el valor crítico de t
t_critico = stats.t.ppf(1 - prob_cola_derecha, df)
print(f"El valor crítico de t con df = {df} que deja una probabilidad de {prob_cola_derecha} en la cola derecha es: {t_critico:.4f}")
```

### Ejercicio 1
Para una $ N(0,1) $. ¿Cuál es el valor de $ P(z \leq 0,5) $?

```python

import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Valor de z
z = 0.5

# Calculando la probabilidad
prob = stats.norm.cdf(z)
print(f"P(z <= {z}) = {prob:.4f}")

# Gráfica
x_vals = np.linspace(-3, 3, 1000)
y_vals = stats.norm.pdf(x_vals)
plt.plot(x_vals, y_vals, label='N(0,1)')
plt.fill_between(x_vals, y_vals, where=(x_vals <= z), color='blue', alpha=0.3, label=f'P(z <= {z})')
plt.title('Distribución Normal Estándar')
plt.xlabel('z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

### Ejercicio 3
Para una $ N(0,1) $. ¿Cuál es el valor de $ P(-0,5 < z \leq 0,5) $?


```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
# Valores de z
z1 = -1.96
z2 = 1.96

# Calculando las probabilidades
prob1 = stats.norm.cdf(z1)
prob2 = stats.norm.cdf(z2)

# Probabilidad del intervalo
prob_intervalo = prob2 - prob1
print(f"P({z1} < z <= {z2}) = {prob_intervalo:.4f}")

# Gráfica
x_vals = np.linspace(-3, 3, 1000)
y_vals = stats.norm.pdf(x_vals)
plt.plot(x_vals, y_vals, label='N(0,1)')
plt.fill_between(x_vals, y_vals, where=((x_vals > z1) & (x_vals <= z2)), color='blue', alpha=0.3, label=f'P({z1} < z <= {z2})')
plt.title('Distribución Normal Estándar')
plt.xlabel('z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

### Ejercicio 4
Para una $ N(1,1) $. ¿Cuál es el valor de $ P(0,5 < x \leq 1,5) $?

```python
# Parámetros de la distribución
mu = 1
sigma = 1

# Valores de x
x1 = 0.5
x2 = 1.5

# Calculando los valores tipificados z
z1 = (x1 - mu) / sigma
z2 = (x2 - mu) / sigma

# Calculando las probabilidades
prob1 = stats.norm.cdf(z1)
prob2 = stats.norm.cdf(z2)

# Probabilidad del intervalo
prob_intervalo = prob2 - prob1
print(f"P({x1} < x <= {x2}) = {prob_intervalo:.4f}")

# Gráfica
x_vals = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y_vals = stats.norm.pdf(x_vals, mu, sigma)
plt.plot(x_vals, y_vals, label=f'N({mu},{sigma**2})')
plt.fill_between(x_vals, y_vals, where=((x_vals > x1) & (x_vals <= x2)), color='blue', alpha=0.3, label=f'P({x1} < x <= {x2})')
plt.title(f'Distribución Normal N({mu}, {sigma**2})')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

### Ejercicio 5
Una línea produce tubos de acero cuya longitud responde a una distribución $ N(100, 20) $. La tolerancia aceptada es 5. Se producen 1000 tubos por hora. ¿Cuántos puede esperarse que sean aceptados por hora?

```python
# Parámetros de la distribución
mu = 100
sigma = 20
tolerancia = 5

# Calculando los límites de aceptación
limite_inferior = mu - tolerancia
limite_superior = mu + tolerancia

# Calculando los valores tipificados z
z1 = (limite_inferior - mu) / sigma
z2 = (limite_superior - mu) / sigma

# Calculando las probabilidades
prob1 = stats.norm.cdf(z1)
prob2 = stats.norm.cdf(z2)

# Probabilidad del intervalo
prob_intervalo = prob2 - prob1

# Calculando el número de tubos aceptados por hora
tubos_aceptados = prob_intervalo * 1000
print(f"Número de tubos aceptados por hora = {tubos_aceptados:.0f}")

# Gráfica
x_vals = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y_vals = stats.norm.pdf(x_vals, mu, sigma)
plt.plot(x_vals, y_vals, label=f'N({mu},{sigma**2})')
plt.fill_between(x_vals, y_vals, where=((x_vals >= limite_inferior) & (x_vals <= limite_superior)), color='blue', alpha=0.3, label=f'Intervalo de aceptación ({limite_inferior}, {limite_superior})')
plt.title(f'Distribución Normal N({mu}, {sigma**2})')
plt.xlabel('Longitud')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

### Ejercicio 6
Una línea produce tubos de acero cuya longitud responde a una distribución $ N(100, 5) $. La tolerancia aceptada es 5. Se producen 1000 tubos por hora. ¿Cuántos puede esperarse que sean aceptados por hora?


```python
# Parámetros de la distribución
mu = 100
sigma = 5
tolerancia = 5

# Calculando los límites de aceptación
limite_inferior = mu - tolerancia
limite_superior = mu + tolerancia

# Calculando los valores tipificados z
z1 = (limite_inferior - mu) / sigma
z2 = (limite_superior - mu) / sigma

# Calculando las probabilidades
prob1 = stats.norm.cdf(z1)
prob2 = stats.norm.cdf(z2)

# Probabilidad del intervalo
prob_intervalo = prob2 - prob1

# Calculando el número de tubos aceptados por hora
tubos_aceptados = prob_intervalo * 1000
print(f"Número de tubos aceptados por hora = {tubos_aceptados:.0f}")

# Gráfica
x_vals = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y_vals = stats.norm.pdf(x_vals, mu, sigma)
plt.plot(x_vals, y_vals, label=f'N({mu},{sigma**2})')
plt.fill_between(x_vals, y_vals, where=((x_vals >= limite_inferior) & (x_vals <= limite_superior)), color='blue', alpha=0.3, label=f'Intervalo de aceptación ({limite_inferior}, {limite_superior})')
plt.title(f'Distribución Normal N({mu}, {sigma**2})')
plt.xlabel('Longitud')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

Se desea obtener el valor de K de forma que para una $ N(0,1) $ $ P(z \leq K) = 0,9970 $.

### Ejercicio 8
Las notas de un examen han tenido una distribución $ N(5.2, 0.3) $. ¿Qué porcentaje de alumnos es razonable que hayan suspendido?

```python
# Parámetros de la distribución
mu = 5.2
sigma = 0.3

# Nota mínima para aprobar
nota_min_aprobar = 5.0

# Calculando el valor tipificado z
z = (nota_min_aprobar - mu) / sigma

# Calculando la probabilidad de suspender
prob_suspenso = stats.norm.cdf(z)
porcentaje_suspenso = prob_suspenso * 100
print(f"Porcentaje de alumnos que han suspendido: {porcentaje_suspenso:.2f}%")

# Gráfica
x_vals = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y_vals = stats.norm.pdf(x_vals, mu, sigma)
plt.plot(x_vals, y_vals, label=f'N({mu},{sigma**2})')
plt.fill_between(x_vals, y_vals, where=(x_vals <= nota_min_aprobar), color='blue', alpha=0.3, label=f'P(x <= {nota_min_aprobar})')
plt.title(f'Distribución Normal N({mu}, {sigma**2})')
plt.xlabel('Notas')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

### Ejercicio 9
Las notas de un examen han tenido una distribución $ N(5.2, 3) $. ¿Qué porcentaje de alumnos es razonable que haya aprobado?

```python
# Parámetros de la distribución
mu = 5.2
sigma = 3

# Nota mínima para aprobar
nota_min_aprobar = 5.0

# Calculando el valor tipificado z
z = (nota_min_aprobar - mu) / sigma

# Calculando la probabilidad de aprobar
prob_aprobado = 1 - stats.norm.cdf(z)
porcentaje_aprobado = prob_aprobado * 100
print(f"Porcentaje de alumnos que han aprobado: {porcentaje_aprobado:.2f}%")

# Gráfica
x_vals = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y_vals = stats.norm.pdf(x_vals, mu, sigma)
plt.plot(x_vals, y_vals, label=f'N({mu},{sigma**2})')
plt.fill_between(x_vals, y_vals, where=(x_vals > nota_min_aprobar), color='blue', alpha=0.3, label=f'P(x > {nota_min_aprobar})')
plt.title(f'Distribución Normal N({mu}, {sigma**2})')
plt.xlabel('Notas')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```
