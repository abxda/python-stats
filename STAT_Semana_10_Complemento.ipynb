```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros para el ejemplo didáctico
z_value = 2.5  # Estadístico Z obtenido de los datos
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)  # Función de densidad de la distribución normal estándar

# Cálculo del valor p para una prueba de dos colas
p_value_two_tailed = 2 * norm.sf(abs(z_value))

# Configuración de la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal (Z)', color='black')

# Sombreado de las dos colas correspondientes al valor p
plt.fill_between(x, y, where=(x >= z_value), color='skyblue', alpha=0.5,
                 label='Área p (cola derecha)')
plt.fill_between(x, y, where=(x <= -z_value), color='skyblue', alpha=0.5,
                 label='Área p (cola izquierda)')

# Líneas verticales en ±z_value
plt.axvline(z_value, color='red', linestyle='--', label=f'Estadístico Z = {z_value}')
plt.axvline(-z_value, color='red', linestyle='--')

# Agregar fórmulas al gráfico usando anotaciones en coordenadas relativas
plt.annotate(r'$Z = \frac{\bar{x} - \mu_0}{\sigma/\sqrt{n}}$',
             xy=(0.05, 0.90), xycoords='axes fraction', fontsize=12,
             bbox=dict(facecolor='white', alpha=0.5))
plt.annotate(r'$p = 2\left(1 - \Phi\left(|Z|\right)\right)$',
             xy=(0.05, 0.83), xycoords='axes fraction', fontsize=12,
             bbox=dict(facecolor='white', alpha=0.5))

plt.title('Relación entre el Estadístico Z y el Valor p en una Prueba de Dos Colas')
plt.xlabel('Valor de Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend(loc='upper center')
plt.grid(True)
plt.show()

print(f"Valor p (dos colas): {p_value_two_tailed:.4f}")
```
